```python
%%writefile /content/openpose_mediapipe.py

import cv2
import json
import numpy as np
import os
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision

# Use the new MediaPipe API (0.10.x)
mp_drawing = mp.solutions.drawing_utils
mp_pose_legacy = mp.solutions.pose

MEDIAPIPE_TO_OPENPOSE = {
    0: 0,
    11: 5, 12: 2,
    13: 6, 14: 3,
    15: 7, 16: 4,
    23: 11, 24: 8,
    25: 12, 26: 9,
    27: 13, 28: 10,
}

def process_image(image_path, json_output_dir, img_output_dir):
    img = cv2.imread(image_path)
    if img is None:
        print(f"Could not read {image_path}")
        return

    h, w = img.shape[:2]
    base_name = os.path.splitext(os.path.basename(image_path))[0]

    with mp_pose_legacy.Pose(
        static_image_mode=True,
        model_complexity=1,
        min_detection_confidence=0.5
    ) as pose:
        results = pose.process(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

        keypoints = []
        if results.pose_landmarks:
            landmarks = results.pose_landmarks.landmark
            for i in range(25):
                mp_idx = next(
                    (k for k, v in MEDIAPIPE_TO_OPENPOSE.items() if v == i),
                    None
                )
                if mp_idx is not None and mp_idx < len(landmarks):
                    lm = landmarks[mp_idx]
                    keypoints.extend([lm.x * w, lm.y * h, lm.visibility])
                else:
                    keypoints.extend([0.0, 0.0, 0.0])
        else:
            print(f"⚠️ No pose detected in {base_name}")
            keypoints = [0.0] * 75

        pose_data = {
            "version": 1.3,
            "people": [{
                "person_id": [-1],
                "pose_keypoints_2d": keypoints,
                "face_keypoints_2d": [],
                "hand_left_keypoints_2d": [],
                "hand_right_keypoints_2d": [],
                "pose_keypoints_3d": [],
                "face_keypoints_3d": [],
                "hand_left_keypoints_3d": []
            }]
        }

        json_path = os.path.join(json_output_dir, f"{base_name}_keypoints.json")
        with open(json_path, "w") as f:
            json.dump(pose_data, f)

        rendered = img.copy()
        if results.pose_landmarks:
            mp_drawing.draw_landmarks(
                rendered,
                results.pose_landmarks,
                mp_pose_legacy.POSE_CONNECTIONS
            )

        rendered_path = os.path.join(img_output_dir, f"{base_name}_rendered.png")
        cv2.imwrite(rendered_path, rendered)
        print(f"✅ Processed {base_name}")

def run_on_directory(image_dir, json_output_dir, img_output_dir):
    os.makedirs(json_output_dir, exist_ok=True)
    os.makedirs(img_output_dir, exist_ok=True)
    for fname in os.listdir(image_dir):
        if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
            process_image(
                os.path.join(image_dir, fname),
                json_output_dir,
                img_output_dir
            )

if __name__ == "__main__":
    import sys
    run_on_directory(sys.argv[1], sys.argv[2], sys.argv[3])
```

```python
%%writefile /content/human_parse_modern.py
"""
Drop-in replacement for Self-Correction-Human-Parsing
"""
import os
import sys
import numpy as np
from PIL import Image
import torch
from transformers import SegformerImageProcessor, SegformerForSemanticSegmentation

def parse_image(image_path, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    img = Image.open(image_path).convert("RGB")
    base_name = os.path.splitext(os.path.basename(image_path))[0]

    print(f"Processing {base_name}...")

    processor = SegformerImageProcessor.from_pretrained("mattmdjaga/segformer_b2_clothes")
    model = SegformerForSemanticSegmentation.from_pretrained("mattmdjaga/segformer_b2_clothes")
    model.eval()

    inputs = processor(images=img, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)

    # Resize to EXACTLY (768 width, 1024 height) matching load_width x load_height
    target_h, target_w = 1024, 768

    upsampled = torch.nn.functional.interpolate(
        outputs.logits,
        size=(target_h, target_w),  # (H, W)
        mode="bilinear",
        align_corners=False
    )
    seg_map = upsampled.argmax(dim=1).squeeze().numpy().astype(np.uint8)

    # Remap to LIP format
    remapping = {
        0: 0, 1: 1, 2: 2, 3: 4, 4: 5,
        5: 12, 6: 9, 7: 6, 8: 0, 9: 18,
        10: 19, 11: 13, 12: 16, 13: 17,
        14: 14, 15: 15, 16: 0, 17: 11,
    }

    lip_map = np.zeros_like(seg_map)
    for src, dst in remapping.items():
        lip_map[seg_map == src] = dst

    # Save — shape is already (1024, 768) = (H, W)
    output_img = Image.fromarray(lip_map, mode='L')
    print(f"Parse map shape: {lip_map.shape} (H, W)")

    output_path = os.path.join(output_dir, f"{base_name}.png")
    output_img.save(output_path)
    print(f"✅ Saved: {output_path}")

def run_on_directory(input_dir, output_dir):
    for fname in os.listdir(input_dir):
        if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
            parse_image(os.path.join(input_dir, fname), output_dir)

if __name__ == "__main__":
    input_dir = sys.argv[1] if len(sys.argv) > 1 else "/content/inputs/test/image"
    output_dir = sys.argv[2] if len(sys.argv) > 2 else "/content/inputs/test/image-parse"
    run_on_directory(input_dir, output_dir)
```

```python
import os, io, subprocess, nest_asyncio
from PIL import Image
from fastapi import FastAPI, File, UploadFile
from fastapi.responses import FileResponse, JSONResponse
import uvicorn
from pyngrok import ngrok

nest_asyncio.apply()

def make_dir():
    for d in ["cloth","cloth-mask","image","image-parse","openpose-img","openpose-json"]:
        os.makedirs(f"/content/inputs/test/{d}", exist_ok=True)

app = FastAPI()

@app.get("/health")
def health():
    return {"status": "ok"}

@app.post("/api/transform")
async def transform(
    model: UploadFile = File(...),
    cloth: UploadFile = File(...)
):
    os.system("rm -rf /content/inputs /content/output")
    make_dir()

    cloth_img = Image.open(io.BytesIO(await cloth.read())).convert("RGB")
    model_img = Image.open(io.BytesIO(await model.read())).convert("RGB")

    cloth_img = cloth_img.resize((768, 1024), Image.BICUBIC)
    model_img = model_img.resize((768, 1024), Image.BICUBIC)

    cloth_img.save("/content/inputs/test/cloth/cloth.png")
    model_img.save("/content/inputs/test/image/model.png")

    with open("/content/inputs/test_pairs.txt", "w") as f:
        f.write("model.png cloth.png")

    # Step 1 - Cloth mask
    os.chdir("/content/clothes-virtual-try-on")
    r1 = subprocess.run(["python", "cloth-mask.py"],
                        capture_output=True, text=True,
                        cwd="/content/clothes-virtual-try-on")
    print("MASK:", r1.stdout[-200:], r1.stderr[-100:])

    # Step 2 - Human parsing (modern replacement)
    r2 = subprocess.run(
        ["python", "/content/human_parse_modern.py",
         "/content/inputs/test/image",
         "/content/inputs/test/image-parse"],
        capture_output=True, text=True
    )
    print("PARSE:", r2.stdout, r2.stderr[-200:])

    # Step 3 - Pose estimation (mediapipe)
    r3 = subprocess.run(
        ["python", "/content/openpose_mediapipe.py",
         "/content/inputs/test/image",
         "/content/inputs/test/openpose-json",
         "/content/inputs/test/openpose-img"],
        capture_output=True, text=True
    )
    print("POSE:", r3.stdout, r3.stderr[-100:])

    # Verify all preprocessing outputs
    print("cloth-mask:", os.listdir("/content/inputs/test/cloth-mask"))
    print("image-parse:", os.listdir("/content/inputs/test/image-parse"))
    print("openpose-img:", os.listdir("/content/inputs/test/openpose-img"))
    print("openpose-json:", os.listdir("/content/inputs/test/openpose-json"))

    # Step 4 - Run ViTON
    r4 = subprocess.run(
        ["python", "/content/clothes-virtual-try-on/run.py"],
        capture_output=True, text=True,
        cwd="/content/clothes-virtual-try-on"
    )
    print("VITON STDOUT:", r4.stdout[-1000:])
    print("VITON STDERR:", r4.stderr[-500:])

    if not os.path.exists("/content/output") or not os.listdir("/content/output"):
        return JSONResponse(
            {"error": f"Pipeline failed: {r4.stderr[-300:]}"},
            status_code=500
        )

    output_file = os.listdir("/content/output")[0]
    return FileResponse(
        f"/content/output/{output_file}",
        media_type="image/png"
    )

ngrok.set_auth_token("3AB0txXJ3yoLnhwTqQXpoz5ag7w_2sVLqRdYaDpQsui1hFC5q")
tunnel = ngrok.connect(8000)
print("="*50)
print(f"✅ API: {tunnel.public_url}")
print(f"🚀 Endpoint: {tunnel.public_url}/api/transform")
print("="*50)

config = uvicorn.Config(app, host="0.0.0.0", port=8000)
server = uvicorn.Server(config)
await server.serve()
```
